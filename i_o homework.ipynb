```python
# Input the names and ages of two people
name1 = input("Enter the name of the first person: ")
age1 = int(input("Enter the age of the first person: "))

name2 = input("Enter the name of the second person: ")
age2 = int(input("Enter the age of the second person: "))

# Print out the output
print("First person:")
print("Name:", name1)
print("Age:", age1)

print("\nSecond person:")
print("Name:", name2)
print("Age:", age2)
```

```python
# Input the numbers
num_list = input("Enter a list of numbers (separated by space): ").split()

# Convert the input numbers to integers
num_list = [int(num) for num in num_list]

# Calculate the average
average = sum(num_list) / len(num_list)

# Print the average
print("The average is:", average)
```

```python
# Input the dimensions of the rectangle
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

# Calculate the area
area = length * width

# Print the area
print("The area of the rectangle is:", area)
```

```python
# Input the limit
limit = int(input("Enter the limit: "))

# Initialize sum variable
sum_of_numbers = 0

# Calculate the sum of natural numbers
for num in range(1, limit + 1):
    sum_of_numbers += num

# Print the sum
print("The sum of natural numbers up to", limit, "is:", sum_of_numbers)
```

```python
import math

# Input the numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Calculate the HCF
hcf = math.gcd(num1, num2)

# Print the HCF
print("The HCF of", num1, "and", num2, "is:", hcf)
```

```python
# Input the number
num = int(input("Enter a number: "))

# Initialize factorial variable
factorial = 1

# Calculate the factorial
if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    for i in range(1, num + 1):
        factorial *= i

    # Print the factorial
    print("The factorial of", num, "is:", factorial)
```

```python
def pypart(n):
    for i in range(0, n):
        for j in range(0, i+1):
            print("* ",end="")
        print("\r")
n = 10
pypart(n)
```
